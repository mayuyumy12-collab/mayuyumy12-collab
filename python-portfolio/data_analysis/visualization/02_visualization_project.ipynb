**Analyzing Factors Influencing Tipping Behavior with Seaborn**\
This notebook esplores tipping patterns using data visualization techniques.

**Introduction**\
The goal of this analysis is to explore the relationships between variables in the dataset and identify key factors that influence tipping behavior.

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.set_theme(style="whitegrid")

tips = sns.load_dataset("tips")

# Pairplot(to see the pairwise relationships across the entire dataset)
pair_plot = sns.pairplot(tips, hue='sex', palette='Set2', diag_kind='kde')
pair_plot.fig.suptitle("Pairwise Relationships by Gender", y=1.02)
plt.show()
```

**Insight for the Pairplot**\
The pairplot reveals a strong positive linear relationship between total_bill and tip. Additionally, according to the KDE plots on the diagonal, most customers tend to visit in groups of two and more.

```python
# Regplot(to visualize the specific correlation between Total Bill and Tip)
plt.figure(figsize=(8, 6))
sns.regplot(data=tips, x="total_bill", y="tip", scatter_kws={'alpha':0.5}, line_kws={'color':'red'})
plt.title("Correlation between Total Bill and Tip Amount")
plt.xlabel("Total Bill ($)")
plt.ylabel("Tip ($)")
plt.show()
```

**Insight for the Regression plot**\
The regression line confirms that tip amounts generally rise with the total bill. However, we observe more spread as the bill increases. This may be because the tip has a smaller impactbon the total cost for customers who spend more.

```python
# Heatmap(to identify the correlation strength between numerical variables)
plt.figure(figsize=(8, 6))
numeric_tips = tips.select_dtypes(include=['float64', 'int64'])  # Select numeric columns only
corr = numeric_tips.corr()
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title("Correlation Matrix Heatmap")
plt.show()
```

**Insight for the Heatmap**\
The heatmap shows a high correlation coefficient between total_bill and tip. On the other hand, the correlation between size and tip is the lowest of the three. This indicates that total_bill has a greater influence on the tip than the group size.
