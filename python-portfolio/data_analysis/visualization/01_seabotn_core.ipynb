**Heatmap**\
Visualize the ice cream sales for each day of the week with a heatmap

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

# the ice cream data by day of the week
data={
    'day':['Sun', 'Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat']*4,
    'sales':[80, 40, 55, 65, 50, 60, 73,
             70, 40, 50, 60, 60, 60, 73,
             82, 45, 52, 67, 66, 70, 75,
             84, 50, 60, 65, 70, 75, 80]
}

df=pd.DataFrame(data)

# assign week numbers
df['week']=[1, 1, 1, 1, 1, 1, 1,   # Week1
            2, 2, 2, 2, 2, 2, 2,   # Week2
            3, 3, 3, 3, 3, 3, 3,   # Week3
            4, 4, 4, 4, 4, 4, 4,]  # Week4

# specify the order of the days of the week
days_order=['Sun', 'Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat']
df['day']=pd.Categorical(df['day'], categories=days_order, ordered=True)

# create a pivot table to restructure the data for the heatmap
pivot_df=df.pivot(index='week', columns='day', values='sales')

# create the heatmap to visualize the sales data
plt.figure(figsize=(6, 4))
sns.heatmap(pivot_df, annot=True, fmt="d", cmap="YlGnBu")
plt.title('Ice Cream Sales')
plt.xlabel('Day of the Week')
plt.ylabel('Week')
plt.show()
```

Sales are higher on weekends and lower at the beginning of the week. This pattern is consist across all weeks.

**Regression Plot**\
explore the relationships among the three groups using regression plot

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

# prepare the actual data
data={
    'group':['A', 'A', 'A', 'A', 'A', 'A', 'A', 'A', 'A', 'A',
             'B', 'B', 'B', 'B', 'B', 'B', 'B', 'B', 'B', 'B',
             'C', 'C', 'C', 'C', 'C', 'C', 'C', 'C', 'C', 'C'],
    'height':[150, 153, 157, 161, 165, 168, 172, 174, 175, 176,
              155, 162, 165, 168, 170, 171, 173, 175, 177, 180,
              160, 163, 166, 168, 170, 173, 175, 178, 180, 183],
    'weight':[58, 57, 55, 62, 58, 65, 62, 78, 70, 72,
              55, 60, 62, 70, 66, 70, 71, 70, 73, 75,
              50, 50, 48, 60, 65, 75, 66, 71, 75, 85]
}

df=pd.DataFrame(data)

# create regression plots for each group
sns.lmplot(data=df, x='height', y='weight', hue='group', palette='deep', markers=['o', 's', '^'], aspect=1.5)

plt.title('Height vs. Weight with Regression Lines by Group')
plt.xlabel('Height(cm)')
plt.ylabel('Weight(kg)')
plt.grid(True)
plt.show()
```

A positive correlation between height and weight is observed in all groups. In particular, Group C has a stronger relationship compared to the other groups.
