**Visualization**\
Read the data from the file and visualize it using Matplotlib

```python
import pandas as pd
from google.colab import files
import matplotlib.pyplot as plt

uploaded=files.upload()
file_name=list(uploaded.keys())[0]
df=pd.read_csv(file_name)
print(df)

fig=plt.figure(figsize=(5, 4))
ax1=fig.add_subplot(1, 1, 1)
ax1.bar(df['Week'], df['Sales'], label='Sales')
ax2=ax1.twinx()  # overlay tables
ax2.plot(df['Week'], df['Temperature'], marker='x', color='red', label='Temperature')
plt.title('Ice Cream Sales and Temperature', fontsize=16)
ax1.set_xlabel('Week', fontsize=14)
ax1.set_ylabel('Sales', fontsize=14)
ax2.set_ylabel('Temperature', fontsize=14)

ax1.set_ylim(0, 100)
ax2.set_ylim(20, 35)

# combine the legend entries from both axes into a single unified legend
lines1, labels1=ax1.get_legend_handles_labels()
lines2, labels2=ax2.get_legend_handles_labels()
ax1.legend(lines1+lines2, labels1+labels2, loc='upper left')
plt.show()
```
