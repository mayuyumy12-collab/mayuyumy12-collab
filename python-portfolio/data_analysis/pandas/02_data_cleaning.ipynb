**Data Cleaning**\
Impute missing values and extract rows that meet the condition

```python
import pandas as pd

data={
    'Name':['Alice', 'Bob', 'Nancy', 'David', 'Eva'],
    'Age':[25, 30, None, 22, 35],
    'Salary':[250000, None, 275000, None, 280000],
    'Gender':['Female', 'Male', 'Female', 'Male', 'Female']
}
df=pd.DataFrame(data)

# mean imputation
df['Salary']=df['Salary'].fillna(df['Salary'].mean())
df['Age']=df['Age'].fillna(df['Age'].mean())

# extract only the rows where the Gender column is Female
df_female=df[df['Gender']=='Female'].copy()

# convert the Salary and Age columns to integers
df_female['Salary']=df_female['Salary'].astype(int)
df_female['Age']=df_female['Age'].astype(int)

print(df_female)
```
