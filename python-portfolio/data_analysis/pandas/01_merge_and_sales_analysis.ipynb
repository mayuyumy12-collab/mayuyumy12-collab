**Data Merging and Computation**\
Combine two tables and perform calculations using their values

```python
import pandas as pd

df_A=pd.DataFrame({
    'Product Number':[1001, 1002, 1003],
    'Product Name':['apple', 'orange', 'banana'],
    'Unit Price':[510, 350, 150]
})

df_B=pd.DataFrame({
    'Product Number':[1001, 1002, 1003],
    'Units Sold':[10, 20, 15]
})

# Merge Table A and Table B
df_M=pd.merge(df_A, df_B, on='Product Number')

# Calculate the sales amount
df_M['Sales Amount']=df_M['Unit Price']*df_M['Units Sold']

print(' Product Number   Product Name     Unit Price     Units Sold   Sales Amount ')
for number, name, price, sales, amount in df_M.itertuples(index=False):
  print(f'{number:15d}{name:>15s}{price:15d}{sales:15d}{amount:15d}')
```
